```python
import random
import struct
import numpy as np
import sys
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from solution import generate_random_points_in_triangle
```

```python
# Three triangle vertexes, and basis vectors
p1, p2, p3 = np.array([1, 0]), np.array([2, 3]), np.array([0, 6])

# Create random samples
random_numbers = np.random.rand(10000)

# Calculate points coordinates
points = generate_random_points_in_triangle(random_numbers, (p1, p2, p3))
```

```python
fig, ax = plt.subplots()
ax.scatter(*zip(p1, p2, p3))
ax.scatter(points[:, 0], points[:, 1], color='red', s=0.1, marker='o')
```

```python
# KDE
```

```python
kernel = scipy.stats.gaussian_kde(points.T, bw_method=0.2)
X, Y = np.mgrid[0:2:100j, 0:6:100j]
Z = kernel(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)
```

```python
fig, ax = plt.subplots()
ax.pcolormesh(X, Y, Z, shading='gouraud')
# ax.scatter(points[:, 0], points[:, 1], color='white', s=0.1, marker='o')
```
